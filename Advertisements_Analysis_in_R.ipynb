#Support Vector Machine

```r
clean_dataset = select(clean_dataset,-c(Ad.Topic.Line,Timestamp,City,Country,outlier,Male))
```

```r
clean_dataset
```

Encoding the target feature as factor

```r
clean_dataset$Clicked.on.Ad = factor(clean_dataset$Clicked.on.Ad, levels = c(0,1))
```

Spliting the dataset into Training set and Test set

```r
install.packages('caTools')
library(caTools)
```

```r
set.seed(123)
split = sample.split(clean_dataset$Clicked.on.Ad, SplitRatio = 0.75)
training_set = subset(clean_dataset, split == TRUE)
test_set = subset(clean_dataset, split == FALSE)
```

Feature Scaling

```r
training_set[1:4] = scale(training_set[1:4])
test_set[1:4] = scale(test_set[1:4])
```

```r
training_set[1:4]
```

```r
test_set[1:4]
```

Fitting Classifier to the Training set

```r
install.packages('e1071')
library(e1071)
```

```r
classifier = svm(formula = Clicked.on.Ad ~ .,
data = training_set, type = 'C-classification', kernel = 'linear')
```

Predicting the Test set results

```r
y_pred = predict(classifier, newdata = test_set[1:4])
y_pred
```

Making the Confusion Matrix

```r
install.packages('caret')
library(caret)
```

```r
confusionMatrix(test_set[, 5], y_pred)
```

* Using  Support Vector Machine we are able to build a model that can successfully predict whether a user will click on an Ad or not with an accuracy of 95%
